# Cào IMDBID

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import re

# Khởi tạo trình duyệt
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Truy cập vào trang JustWatch của phim Deadpool 3
url = "https://www.justwatch.com/us/movie/deadpool-3"
driver.get(url)

# Lấy nội dung trang
page_source = driver.page_source

# Tìm imdbId trong JSON
imdb_id_match = re.search(r'"imdbId":"(tt\d+)"', page_source)
if imdb_id_match:
    imdb_id = imdb_id_match.group(1)
    print("IMDB ID:", imdb_id)
else:
    print("Không tìm thấy IMDb ID.")

# Đóng trình duyệt
driver.quit()
```

# Cuộn và Cào

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import re

# Khởi tạo trình duyệt
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Truy cập trang JustWatch
url = "https://www.justwatch.com/us?age_certifications=G,NC-17,PG,PG-13,R,UNRATED&release_year_from=2023&release_year_until=2024&sort_by=release_year"
driver.get(url)

# Cấu hình thời gian chờ giữa mỗi lần cuộn
scroll_pause_time = 2
seen_movies = set()  # Tập hợp để lưu trữ liên kết đã cào

# Mở file để ghi imdbId
with open("links.txt", "a") as file:

    # # Hàm lấy thông tin phim từ tab mới
    # def scrape_movie_info(movie_url):
    #     driver.execute_script("window.open(arguments[0]);", movie_url)
    #     driver.switch_to.window(driver.window_handles[-1])
    #     time.sleep(4)

    #     try:
    #         # Lấy IMDb ID từ JSON trong source của trang
    #         page_source = driver.page_source
    #         imdb_id_match = re.search(r'"imdbId":"(tt\d+)"', page_source)
    #         if imdb_id_match:
    #             imdb_id = imdb_id_match.group(1)

    #             # Lưu vào danh sách và ghi vào file
    #             imdb_ids.append(imdb_id)
    #             file.write(imdb_id + "\n")  # Ghi ngay vào file
    #             file.flush()  # Đảm bảo ghi vào đĩa
    #     except Exception as e:
    #         print("Lỗi khi lấy thông tin:", e)

    #     # Đóng tab và quay lại tab chính
    #     driver.close()
    #     driver.switch_to.window(driver.window_handles[0])

    # Cuộn và cào dữ liệu theo từng phần
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Lấy các liên kết phim mới sau khi cuộn
        movie_links = driver.find_elements(By.CSS_SELECTOR, "a.title-list-grid__item--link")

        for link in movie_links:
            movie_url = link.get_attribute("href")
            if movie_url not in seen_movies:
                # scrape_movie_info(movie_url)
                seen_movies.add(movie_url)
                
                # Lưu vào danh sách và ghi vào file
                file.write(movie_url + "\n")  # Ghi ngay vào file
                file.flush()  # Đảm bảo ghi vào đĩa

        # Cuộn xuống cuối trang
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause_time)

        # Kiểm tra chiều cao của trang, nếu không thay đổi thì đã đến cuối trang
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        print(len(seen_movies))

# Đóng trình duyệt
driver.quit()
```
